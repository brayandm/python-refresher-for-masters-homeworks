#### Exercise 3.9: Extract parent objects from the dataset

```python
last_object = None

def get_last_object(x):
  global last_object
  if not x.endswith(")"):
    last_object = x

  return last_object.lower()

university_df['parent_object_raw'] = university_df['raw'].apply(get_last_object)
university_df.head(40)
```

#### Exercise 3.10: Extract lisp data from it


it should look like this:
image.png

```python
def getA(x):
    if x.strip().startswith('('):
      return x.strip().split()[0][1:].lower()

def getB(x):
    if x.strip().startswith('('):
      val =  ' '.join(x.strip().split()[1:])
      return val[:-1] if val.endswith(')') else val

university_df['lisp_A'] = university_df['raw'].apply(getA)
university_df['lisp_B'] = university_df['raw'].apply(getB)

university_df = university_df.dropna(subset=['lisp_A'])

university_df.head(20)
```

#### Exercise 3.11: Extract subdatasets for the universities

```python
subdatasets = {}
for parent_uni_def in university_df.query("lisp_A == 'def-instance'")['parent_object_raw'].unique():
    subdatasets[parent_uni_def] = university_df.query("parent_object_raw == '%s'" % parent_uni_def)

subdatasets['(def-instance adelphi']
```

#### Exercise 3.12*: Merge dataset for all universities

```python
cumulative_bringed = pd.DataFrame(columns = [])

for parent_uni_def, subdataset in subdatasets.items():
    subdataset_copy = subdataset.copy()
    subdataset_copy['PivotIndex'] = 0
    df_pivot = subdataset_copy.pivot_table(index='PivotIndex', columns='lisp_A', values='lisp_B', aggfunc='first')
    df_pivot.reset_index(drop=True, inplace=True)
    cumulative_bringed = pd.concat([cumulative_bringed, df_pivot], axis=0, ignore_index=True)

cumulative_bringed
```

#### Exersice 3.21:  Do classification using linear regression
- split data to train/test
- create Pipeline
- do prediction
- compare with answers


```python
from sklearn.metrics import mean_squared_error

X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2, random_state=1)

pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('regressor', LinearRegression())
])

pipeline.fit(X_train, y_train)

predictions = pipeline.predict(X_test)

# Since LinearRegression outputs continuous values, round them to get discrete class predictions
predictions = np.rint(predictions)

mse = mean_squared_error(y_test, predictions)
rmse = np.sqrt(mean_squared_error(y_test, predictions))
accuracy = np.mean(predictions == y_test)

print("Predictions:", predictions)
print("MSE:", mse)
print("RMSE:", rmse)
print("Accuracy:", accuracy)
```

#### Exersice 3.22*:  Do classification using KNN
- split data to train/test
- create Pipeline
- do prediction (find what you can classify in the dataset)
- use [metrics](https://scikit-learn.org/stable/modules/model_evaluation.html) to check your result
- try to make a better prediction

```python

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2, random_state=1)

pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('knn', KNeighborsClassifier(n_neighbors=3))
])

pipeline.fit(X_train, y_train)

predictions = pipeline.predict(X_test)

accuracy = accuracy_score(y_test, predictions)
mse = mean_squared_error(y_test, predictions)
rmse = np.sqrt(mean_squared_error(y_test, predictions))
accuracy = np.mean(predictions == y_test)

conf_matrix = confusion_matrix(y_test, predictions)
report = classification_report(y_test, predictions)

print("MSE:", mse)
print("RMSE:", rmse)
print("Accuracy:", accuracy)
print("Confusion Matrix:\n", conf_matrix)
print("Classification Report:\n", report)
```

#### Exercise 8.2* Draw the segmentation map for apple

```python
def our_apple_segmentation_function(x):
  if x[0] > 140.0:
    return 0.5 + (x[0]-140)/115.
  else:
    return 0.
plt.imshow(np.apply_along_axis(our_apple_segmentation_function, 2,
                               im), cmap=plt.get_cmap('gray'))
```

#### Exercise 8.3*: Create a simple detector of apple image seminar using numpy, that outputs most probable coordinates of apple on the image. Try using only native to numpy methods.

```python
def detect_blankish_center(image, window_width, window_height, steps):
    # Convert the image to grayscale
    gray_image = np.apply_along_axis(our_apple_segmentation_function, 2,
                               im)

    rows, cols = gray_image.shape
    max_blankness = 0
    best_center = (0, 0)

    # Slide the window over the image
    for y in range(0, rows - window_height + 1, steps):
        for x in range(0, cols - window_width + 1, steps):
            # Compute the blankness of the current window
            window = gray_image[y:y+window_height, x:x+window_width]
            blankness = np.mean(window)

            # Update the maximum blankness and the best center
            if blankness > max_blankness:
                max_blankness = blankness
                best_center = (x + window_width // 2, y + window_height // 2)

    return best_center

center = detect_blankish_center(im, 150, 150, 10)

print(center)

dot_size = 50
half_size = dot_size // 2

new_im = im.copy()

new_im[max(0, center[1]-half_size):min(im.shape[0], center[1]+half_size),
   max(0, center[0]-half_size):min(im.shape[1], center[0]+half_size), :] = [255, 255, 0]

# Display the image
plt.imshow(new_im)
plt.title("Image with Yellow Dot")
plt.show()
```
